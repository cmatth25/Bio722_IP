```python
#set index to sample to keep column when dividing columns with matching samples in different dataframes
longtrimpair.set_index('Sample', inplace=True)
long.set_index('Sample', inplace=True)
```

```python
#determine paired sequences trimmed with sufficient quality
trim_prop = longtrimpair["Total Sequences"].div(long["Total Sequences"]).drop_duplicates().reset_index()
trim_prop.columns = ['Sample','paired Sequences passed QC']
trim_prop
```

```python
#summary from unpaired reads passing filters
longtrimunpair = pd.read_csv(path+file6, sep = '\t')
longtrimunpair
```

```python
#split sample ID
longtrimunpair['Sample'] = longtrimunpair['Sample'].str.split("_", expand = True)[0]
#split read direction
longtrimunpair['dir'] = longtrimunpair['Filename'].str.split("_", expand = True)[1]
longtrimunpair
```

```python
#set index for division
longtrimunpair.set_index('Sample', inplace=True)
longtrimunpair
```

```python
#divide, include column for direction
uptrim_prop = longtrimunpair["Total Sequences"].div(long["Total Sequences"]).reset_index()
uptrim_prop.columns = ['Sample','unpaired Sequences passed QC']
uptrim_prop['dir'] = longtrimunpair['dir'].values
uptrim_prop
```

```python
#merge trim and untrimmed on sample
prop = pd.merge(uptrim_prop,trim_prop,how = 'left',on=["Sample"])
```

```python
#combine paired and unpaired for each direction
prop['Total Sequences passed QC'] = prop['unpaired Sequences passed QC'] + prop['paired Sequences passed QC'] 
```
