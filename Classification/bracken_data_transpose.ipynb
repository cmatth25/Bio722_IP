```python
# import libraries
import numpy as np
import pandas as pd
import json
import sys
import matplotlib

#load processed kraken formatted bracken report
phylname = "process_bracken.breport"
path = "C:/Users/cmatt/Dropbox/Craig/Bio722/breport/" #PCdirectory path
#path = "/Users/craigmatthews/Dropbox/Craig/Bio722/breport/" #mac
phyl = path + phylname
pbracken = pd.read_csv(phyl, sep='\\s+')
pbracken
```

```python
#take a closer look
pbracken.head(50)
```

```python
#check sum rounds to 100, close enough.
Total = pbracken["#perc"].sum()
Total
```

```python
#remove chordata, copy to alter
pbracken = pbracken[~pbracken['name'].isin(["Chordata"])]
pbrackencopy = pbracken
pbracken
```

```python
#create other column for taxa less than 2%%
pbrackencopy["New Index"] = np.where(pbrackencopy["#perc"] < 0.02, "Other", pbrackencopy["#perc"].index)

pbrackencopy['name'] = np.where(pbrackencopy["New Index"].eq("Other") ,"Other",pbrackencopy["name"])

pbrackencopy = pbrackencopy.drop(["New Index", "taxid"], axis =1)
#PC throws error, not on mac
```

```python
pbrackencopy
```

```python
#add Other taxa together
pbrackencopy = pbrackencopy.groupby("name").sum(numeric_only=True).sort_values(by=["#perc"], ascending=False)
pbrackencopy
```

```python
#create copy to normalize to percentage for each sample
norm = pbrackencopy.drop(["#perc","tot_all"], axis=1)
norm
```

```python
#convert counts to percent
column_names = list(norm.columns)
cols = column_names
norm[cols] = norm[cols].div(norm[cols].sum(axis=0), axis=1).multiply(100)
norm
```

```python
#check samples sum to 100%
norm[cols].sum()
```

```python
#quick plot to look at samples, ugly as it may be, it looks okay
norm.transpose().plot.bar(stacked=True)
```

```python
#transpose so filtered taxa are cols, samples are rows
transposed  = norm.transpose().reset_index().rename(columns={'index': 'sample'})
transposed 
```

```python
#convert sample col, dropping _all
transposed['sample'] = transposed['sample'].str.split("_", expand = True)[0]
transposed
```

```python
path2 = "C:/Users/cmatt/Dropbox/Craig/Bio722/" # PC
#path2 = "/Users/craigmatthews/Dropbox/Craig/Bio722/" #maac
eco = pd.read_csv(path2+"SampleEcosystem.csv")
eco
```

```python
#join sample ecosystem type to transposed table
final = pd.merge(transposed,eco,how = 'left',on=["sample"])
final
```
